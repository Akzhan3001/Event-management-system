```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
from catboost import CatBoostRegressor
```

```python
df = pd.read_csv('salary_train.csv')
```

```python
df
```

```python
dp = pd.read_csv('salary_predict.csv')
dp
```

```python
df['job'].unique()
```

# Junior developer

```python
j_d = df[df.job == 'junior developer']
j_d
```

```python
j_d_pred = dp[dp.job == 'junior developer']
j_d_pred_copy=j_d_pred.copy()
j_d_pred
```

```python
j_d.drop(['Id'], axis=1, inplace=True)
j_d.drop(['job'], axis=1, inplace=True)
j_d
```

```python
j_d_pred.drop(['Id'], axis=1, inplace=True)
j_d_pred.drop(['job'], axis=1, inplace=True)
j_d_pred
```

```python
y = j_d['salary']
df = j_d.drop(['salary'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
print("Score:", model.score(X_test, y_test))
```

```python
x_pred = j_d_pred.drop(['salary'], axis=1)
predictions = model.predict(x_pred)
```

```python
final = pd.DataFrame({'Id':j_d_pred_copy['Id'], 'salary': predictions})
final['salary'] = np.where(final['salary']>1000000, 1000000, final['salary'])
final = final.astype({'salary': 'int64'})
final
```
