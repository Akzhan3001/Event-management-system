```python
#import all necessary libraries
def import_libs():
  import requests
  from bs4 import BeautifulSoup as soup
  import pandas as pd


def get_html_from_url(url:str):
  html = requests.get(url).content
  return html

def load_page_and_filter(html):
  soup_page = soup(html,"html")
  shoes = soup_page.findAll("div", {"class":"good-box"})
  return shoes

def generate_df_from_soup(soup):
  names = []
  prices = []
  for shoe in shoes:
    names.append(shoe.a.span.text)
    prices.append(shoe.div.p.text)

  df = pd.DataFrame()
  df['Name'] = names
  df['Price'] = prices
  df['Price']=df['Price'].apply(lambda x: float(x.split('\xa0')[-1]))
  return df

def save_csv(filename,df):
  df.to_csv(filename)


```

```python
def run_web_scraping(url,filename):
  import_libs()
  html = get_html_from_url(url)
  soup = load_page_and_filter(html)
  df = generate_df_from_soup(soup)
  save_csv(filename,df)
  print('CSV file is ready!')
```

```python
run_web_scraping('https://www.adidas.com.hk/men/shoes/basketball','adidas.csv')
```

```python
#Load the HTML content
url = 'https://www.adidas.com.hk/men/shoes/basketball?locale=en_GB'
html = requests.get(url).content
soup_page = soup(html,"html")
shoes = soup_page.findAll("div", {"class":"good-box"})

#Create a Pandas DF from HTML
names = []
prices = []
for shoe in shoes:
  names.append(shoe.a.span.text)
  prices.append(shoe.div.p.text)


adidas_shoes_dict = {
    'Name': names,
    'Price': prices
}
df = pd.DataFrame(adidas_shoes_dict)
df = pd.DataFrame()
df['Price']=df['Price'].apply(lambda x: float(x.split('\xa0')[-1]))

#Save df to CSV file
df.to_csv('adidas_shoes.csv')
```
